# 02. 데이터 클리닝
전처리된 파일의 결측치를 제거하여 최종 산출물을 생성합니다.

```python
import pandas as pd
from google.colab import files
```

```python
df = pd.read_csv('/final_airbnb.csv')

print(f"청소 전 원래 데이터 개수: {len(df)}개")

df_clean = df.dropna()

print(f"빈칸 청소 후 남은 데이터 개수: {len(df_clean)}개")

file_name = 'final_airbnb.csv'
df_clean.to_csv(file_name, index=False)
files.download(file_name)
```
